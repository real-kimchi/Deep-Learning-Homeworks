```python
#problem 2, number of updates until convergence using perceptron algo

import numpy as np

#dataset points
pos_class = [(1,3)]
neg_class = [(-1,4)]
data = [(np.array([1,3]),1),
        (np.array([-1,4]),-1)]

#starting weight vector and iteration count
def perceptron(data):
    w = np.array([0,0])
    s = 0
    weight_updates = []

    #perceptron updating loop
    while True:
        incorrect_predictions = 0
        for x, y in data:
            if y * np.dot(w,x) <= 0:
                w = w+y*x
                s += 1
                weight_updates.append(w.copy())
                incorrect_predictions += 1

        if incorrect_predictions == 0:
            break

    return w,s, weight_updates

final_w, weight_updates, weight_updates = perceptron(data)
```

```python
perceptron(data)
```

```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

sns.set_theme()
sns.set_context("paper")
plt.figure(figsize=(10, 6))
sns.scatterplot(x=[x[0] for x in pos_class], y=[x[1] for x in pos_class], s=100, marker="o", label="Positive Class (+1)")
sns.scatterplot(x=[x[0] for x in neg_class], y=[x[1] for x in neg_class], s=100, marker="x", label="Negative Class (-1)")

x_vals = np.linspace(-2, 2, 100)
y_vals = (-final_w[0] / final_w[1]) * x_vals
plt.plot(x_vals, y_vals, "--k", label="Decision Boundary")

plt.title("Perceptron Convergence")
plt.legend()
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()

final_w, weight_updates
```

The perceptron algorithm converged after 5 updates with final weight vector of [5,1]. The decision boundary line y = -5x shows that both data points are seperated classes in the feature space.
